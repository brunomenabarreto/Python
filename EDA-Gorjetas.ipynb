#Teste de HipÃ³tese

**H<sup>null</sup>** "A distribuÃ§Ã£o da taxa da gorjeta Ã‰ a mesma nos dois grupos"

**H<sup>alt</sup>** "A distribuÃ§Ã£o da taxa da gorjeta NÃƒO Ã© a mesma nos dois grupos"

```python
from scipy.stats import ranksums
```

```python
gorjetas.query("sobremesa == 'Sim'").porcentagem
```

```python
sobremesa = gorjetas.query("sobremesa == 'Sim'").porcentagem
```

```python
gorjetas.query("sobremesa == 'NÃ£o'").porcentagem
sem_sobremesa = gorjetas.query("sobremesa == 'NÃ£o'").porcentagem
```

```python
ranksums(sobremesa, sem_sobremesa)
```

```python
r = ranksums(sobremesa, sem_sobremesa)
print('O valor do p-value Ã© {}'.format(r.pvalue))
```
